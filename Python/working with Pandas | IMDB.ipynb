#Ejercicio 4

Compara la media de estrellas en las películas que duran dos horas o más frente a las que duran menos de dos horas.

Almacena en la variable solucion_4 la media de estrellas mayor, redondeada a dos decimales.

```python
movies_less_2 = movies[movies['duration'] < 120] #Peliculas que duran menos de 2h
movies_more_2 = movies[movies['duration'] >= 120] #Peliculas que duran > 2h

star_mean_less = movies_less_2['star_rating'].mean() #Media de estrellas en películas de menos de 2h
star_mean_more = movies_more_2['star_rating'].mean() #Media de estrellas en películas de menos de 2h

print(star_mean_less)
print(star_mean_more)
```

```python
solucion_4= star_mean_more.round(2)

solucion_4
```

#Ejercicio 5

# Ejercicio 6

Almacena en la variable solucion_6 el dataframe resultante de buscar la pelicula con mayor puntuación de estrellas para cada género.

```python
best_star= movies.groupby('genre')['star_rating'].idxmax()

solucion_6 = movies.loc[best_star, ['genre', 'title', 'star_rating']]

solucion_6
```
