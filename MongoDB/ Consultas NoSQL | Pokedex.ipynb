* ¿Cuantas megas hay?

```python
#Realiza una consulta para comprobar cuantas formas megas hay (que contenga Mega en el name)
list(db.segundo_proyecto.find({'Name':{'$regex':'Mega'}}))
```


* Elimina las formas mega

```python
#Elimina las formas mega de la lista
db.segundo_proyecto.delete_many({'Name':{'$regex':'Mega'}})
#1{}: para filtrar, 2{}: operación de actualización
list(db.segundo_proyecto.find({'Name':{'$regex':'Mega'}}))
```

* ¿Cuál es el pokemon legendario más rápido?


```python
#Realiza una consulta de cual es el pokemon legendario más rápido
list(db.segundo_proyecto.aggregate([
    {'$match': {'Legendary':'True'}},
    {'$sort': {'Speed': -1}},
    {'$limit': 1}
    ]))
```

* ¿Cuál es el pokemon no legendario con total stats más alta?

```python
#Realiza una consulta para saber cual es el pokemon no legendario con mayor total de stats
list(db.segundo_proyecto.aggregate([
    {'$match': {'Legendary':'False'}},
    {'$sort': {'Total': -1}},
    {'$limit': 1}
    ]))
```

* Calcula la media del total de stats

```python
#Realiza una consulta para calcular la media del total de stats
list(db.segundo_proyecto.aggregate([
    {'$group':{'_id':'None','Media':{'$avg':'$Total'}}}
]))
```

* Ordena los pokemon de tipo planta según orden alfabético del nombre

```python
#Realiza una consulta para mostrar ordenados alfabéticamente los pokemon de tipo planta (tanto tipo 1 o 2)
list(db.segundo_proyecto.aggregate([
    {'$match': {'$or': [{'Type 1':'Grass'}, {'Type 2':'Grass'}]}},
    {'$sort': {'Name': 1}}
    ]))
```

El error está en la estructura de salida del último ejercicio.

Por otro , la salida es un diccionario

Te recomiendo que primero:
Almacenes los tipos que existen en una lista ordenada alfabéticamente. Puedes hacer querys de MongoDB con los tipos
Estos tipos almacenarlos en un conjunto de Python (set)
Convertir el conjunto a una lista (list)
Ordenar la lista (sorted).
Eliminar vacíos (.remove(''))

Con eso tendrás una lista con todos los tipos.

Después, por cada tipo en la lista (bucle for):
Añade la cuenta, tanto en type1 como type2 (usa mongodb para contar), en una nueva lista resultados.

Al final te tiene que quedar algo así:

[12, 4, 9, 5, 8, 12, 18, 3, 14, 14, 5, 22, 33, 14, 11, 2, 32]

```python
pre_result = list(db.segundo_proyecto.aggregate([
    #Creamos nuevo documento (project) que se llama 'types' con los 2 tipos:
    {'$project': {'types': {'$setUnion': [['$Type 1','$Type 2']]}}},
    #$unwind para deshacer el array y generar un documento separado x cada tipo.
    {'$unwind': '$types'},
    # Agrupa los documentos según el campo _id
    {'$group': {'_id': '$types', 'Count': {'$sum': 1}}},
    # Filtras por los que no están vacios:
    {'$match': {'_id': {'$ne': ''}}},
    #Ordenas alfabeticamente:
    {'$sort':{'_id':1}},
]))

resultados = [types['Count'] for types in pre_result]
print(resultados)
```
